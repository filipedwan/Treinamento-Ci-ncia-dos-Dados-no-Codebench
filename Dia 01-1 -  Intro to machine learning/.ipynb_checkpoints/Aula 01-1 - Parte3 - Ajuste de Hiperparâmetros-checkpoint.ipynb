## Ajuste de hiperparâmetros com validação cruzada

* Neste exemplo, iremo selecionar os melhores parâmetros para o algoritmo KNN utilizando o iris dataset.
* Iremos separar 30% da base de dados para realizar o teste.
* A otimização de hiperparâmetros será realizada com 70% dos dados (o que sobrou).

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline

iris = load_iris()
X = iris.data
y = iris.target

#separando 30% dos dados para realizar o teste após a otimização.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# buscando o melhor valor de K para o KNN
k_range = list(range(1, 26))
k_scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
    k_scores.append(scores.mean())

# plot the value of K for KNN (x-axis) versus the cross-validated accuracy (y-axis)
plt.figure(figsize=(14,7))
plt.plot(k_range, k_scores)
plt.xlabel('Valor de K para o KNN')
plt.ylabel('Acurácia na validação cruzada')
```

## Ajuste de hiperparâmetros de modo mais eficiente

* O grid search é um método que realiza uma busca exaustiva pelo espaço de busca definido
* No sklearn, ele é usado juntamente com a validação cruzada;
* Para tanto, definimos quais hiperparâmetros devem ser testados e o gridsearch retorna os parâmetros que obtiveram a maior acurácia.

### Hiperparâmetros do KNN

```python
from IPython.display import display, HTML
#HTML('<iframe src=http://scikit-learn.org/stable/index.html width=1000 height=350></iframe>')
HTML('<iframe src=http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html width=1000 height=350></iframe>')
```

```python
from sklearn.model_selection import GridSearchCV
import scipy.stats as st

params = {  
    "n_neighbors": list(range(1, 26)), #valor de k
    "leaf_size": [20, 30, 40], #o padrão é 30
    "metric": ['euclidean', 'manhattan', 'chebyshev', 'minkowski'], #métrica para calcular a distância
}

knn = KNeighborsClassifier()
grid = GridSearchCV(knn, params, cv=10, n_jobs=-1, verbose=1) #n_jobs = -1 faz a computação em paralelo
grid.fit(X_train, y_train)

print(grid.best_params_)
print(grid.best_estimator_)
print(grid.best_score_)
```

## Testando o modelo exportado pelo GridSearch

```python
knn = KNeighborsClassifier(leaf_size= 20, metric= 'chebyshev', n_neighbors= 5)
scores = cross_val_score(knn, X_train, y_train, cv=10, scoring='accuracy')
print(scores.mean())
```

## Validando o modelo exportado com os dados separados no início

* Vamos usar agora os 30% dos dados que foram separados no início deste experimento.

```python
from sklearn.metrics import accuracy_score

y_pred = grid.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(acc)
```

## Random Search - reduzindo o custo computacional

* Dependendo do tamanho do espaço de busca, uma busca exaustiva com o Grid Search pode não ser viável computacionalmente;
* Random Search faz uma busca aleatória em um subconjunto dos parâmetros;
* O usuário define o número de iterações do Random Search

```python
from sklearn.model_selection import RandomizedSearchCV

knn = KNeighborsClassifier()

grid = RandomizedSearchCV(knn, params, cv=10, n_jobs=-1, verbose=1, n_iter=100)
grid.fit(X_train, y_train)

print(grid.best_params_)
print(grid.best_estimator_)
print(grid.best_score_)
```
