Iremos usar os seguintes dados: 
* Jogos da copa do mundo 2018:
    * [FIFA 2018 Statistics](https://www.kaggle.com/mathan/fifa-2018-match-statistics)
    * Atributos:
        * **Date**: data da partida;
        * **Team**: nome da seleção;
        * **Opponent**: nome da seleção oponente;
        * **Goal Scored**: número de gols marcados pelo *Team*;
        * **Ball Possession**: quantidade de tempo que a bola esteve sob o controle do Team;
        * **Attempts**: número de tentativas de gol;
        * **On-Target**: número de chutes no alvo (gol);
        * **Off-Target**: número de chutes para fora;
        * **Blocked**: número de chutes bloqueados pela defesa;
        * **Corners**: número de escanteios;
        * **Offsides**: *número de impedimentos*
        * **Free Kicks** *número de cobranças de falta*
        * **Saves**: número de defesas do goleiro;
        * **Pass Accuracy %**: percentual de passes certos;
        * **PassesTotal**: número total de passes do *Team*
        * **Distance Covered (Kms)**: distância total percorrida pelos jogados do *Team* no jogo;
        * **Fouls Committed**: número de faltas cometidas pelo *Team*
        * **Yellow Card**: número de cartões amarelos recebidos;
        * **Yellow & Red**: número de cartões amarelos seguidos com vermelho;
        * **Red**: número de cartões vermelhos;
        * **Man of the Match**: algum dos jogadores do *Team* receberam o prêmio de *Man of the Match*?
        * **1st Goal**: quando foi marcado o primeiro gol?
        * **Round**: se a partida aconteceu na fase de grupos, oitavas, quartas, semi, disputa de 3º lugar ou na final;
        * **PSO**: Existiu um penalty na partida?
        * **Goals in PSO**: número de gols marcados de penalty;
        * **Own goals**: número de gols contra;
        * **Own goal Time**: momento em que o gol contra foi marcado;

### Selecionando linhas e colunas utilizando os rótulos - função loc
* Iremos criar uma cópia do dataFrame que terá o atributo Date como índice.
* Para tanto, usaremos a função **set_index**
* No pandas, as linhas são rotuladas com os índices.
* Através da função loc podemos acessar os dados utilizano os rótulos das linhas e colunas

```python
df_new = df.set_index('Date')
df_new.loc[['14-06-2018', '15-06-2018'], :] #Teria o mesmo efeito o comando df_new.loc[['14-06-2018', '15-06-2018']]
```

* Antes da vírgula temos a seleção das linhas e depois a seleção das colunas.
* Podemos, entre outras coisas, passar uma lista para seleção ou um range separado por ':'

```python
df_new.loc[['14-06-2018', '15-06-2018'], 'Team':'Blocked']
```

```python
df_new.loc[['14-06-2018', '15-06-2018'], ['Team', 'Blocked']]
```
