# Biblioteca seaborn

```python
import seaborn as sns
```

```python
df.columns
```

```python
sub = ['Goal Scored','Ball Possession %',
       'Attempts', 'On-Target', 'Off-Target', 'Blocked', 'Corners', 'Offsides',
       'Free Kicks', 'Saves', 'Pass Accuracy %', 'Passes',
       'Distance Covered (Kms)', 'Fouls Committed', 'Yellow Card']
temp = df[sub]
# Normalize Tournament Data(Min-Max Scale)
data_norm = (temp - temp.mean()) / (temp.max() - temp.min())
plot1 = sns.boxplot(data = data_norm)
plot1.set_xticklabels(sub, rotation=90, ha="center")
plot1.set(xlabel='Métricas',ylabel='Total Normalizado')
plot1.set_title('Boxsplot de métricas dos jogos da copa do mundo 2018')
```

### Plotando a quantidade tentativas de gol por seleção

```python
attempts=df.groupby('Team')['Attempts'].sum().reset_index().sort_values(by=('Attempts'),ascending=False)
plot1 = sns.barplot(x="Team", y="Attempts", data=attempts)
plot1.set_xticklabels(attempts['Team'], rotation=90, ha="center")
plot1.set(xlabel='Seleções',ylabel='Total de Tentativas de Gol')
plot1.set_title('Número de tentativas de gols por seleção')
```

## Total On-target, Off-target e blocked por seleções

```python
group_attempt = df.groupby('Team')['On-Target','Off-Target','Blocked'].sum().reset_index()

# Modificando o dataframe para plotar os valores
group_attempt_sorted = group_attempt.melt('Team', var_name='Target', value_name='Value')

plot1 = sns.barplot(x="Team", y="Value", hue="Target", data=group_attempt_sorted)
plot1.set_xticklabels(group_attempt_sorted['Team'], rotation=90, ha="center")
plot1.set(xlabel='Seleções',ylabel='Total de Tentativas de Gol')
plot1.set_title('Total On-Target, Off-Target e Blocked')
```

### Plotando a quantidade de vezes que as seleções tiveram um de seus jogadores como Man of the Match

```python
mom=df.groupby('Team')['Man of the Match'].sum().reset_index().sort_values(by=('Man of the Match'),ascending=False)

plot1 = sns.barplot(x="Team", y="Man of the Match", data=mom)
plot1.set_xticklabels(mom['Team'], rotation=90, ha="center")
plot1.set(xlabel='Seleções',ylabel='Total de Prêmios de Man of the Match')
plot1.set_title('Quantidade por seleção de prêmios de Man of the Match')
```
