```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/FIFA_2018_Statistics_sem_NaN_novos_Atributos.csv')
numerical_features   = df.select_dtypes(include = [np.number]).columns
df = df[numerical_features]
df.head()
```

## Abordagem inocente para predição do Man of the Match

```python
X, y =  df.drop('Man of the Match', axis=1).values, df['Man of the Match'].values

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

clf = RandomForestClassifier(n_estimators=500)
cv_score = cross_val_score(clf, X, y, scoring="accuracy", cv=10)
```

```python
print('Acc.: %.2f [+/-%.2f]' % (cv_score.mean(), cv_score.std()))
```

## Abordagem inocente para predição do vencedor da partida

```python
X, y =  df.drop('Winner', axis=1).values, df['Winner'].values
clf = RandomForestClassifier(n_estimators=500)
cv_score = cross_val_score(clf, X, y, scoring="accuracy", cv=10)
```

```python
print('Acc.: %.2f [+/-%.2f]' % (cv_score.mean(), cv_score.std()))
```
