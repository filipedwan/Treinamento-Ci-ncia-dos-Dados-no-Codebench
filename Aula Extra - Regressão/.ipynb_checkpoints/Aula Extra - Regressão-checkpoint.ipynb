```python
import pandas as pd

#ler o CSV e usa a primeira coluna como index
data = pd.read_csv('Advertising.csv', index_col=0)
data.head()
```

### Atributos utilizados

* TV: custo em milhares de dolars com propaganda na TV
* Radio: custo propaganda no Rádio
* Newspaper: custo propaganda em jornais;


### Target

* Sales: venda de um produto em um dado mercado

<br>
Como o target é contínuo, então estamos lidando com um problema de **regressão**.

## Regressão Linear

A regressão linear encontrar uma reta que melhor se encaixe aos dados.
* Prós: rápida, facilmente interpretável;
* Contras: se os dados possuiem uma relação não-linear, então este algoritmo não será útil.    

Forma da Regressão Linear:
<br>  
$ f(x_1, ..., x_n) = \beta_0 + \beta_1x_1 + ... + \beta_nx_n $

Chamaremos $f(x_1, ..., x_n)$ de $y$.
<br>
* $y$ é a resposta (target);
* $\beta_0$ é o coeficiente de intercepção;
* $\beta_1$ é o coeficiente para o $x_1$ (primeiro atributo)
* $\beta_n$ é o coeficiente para o $x_n$ (n-ésimo atributo)

Para a base de dados utilizada, temos:
<br>
$ f(x_1, ..., x_n) = \beta_0 + \beta_1 \times TV + \beta_2 \times Radio + \beta_3 \times Newspaper $

```python
#separando os atributos dos targets
X = data[['TV', 'Radio', 'Newspaper']]
y = data['Sales']

#divisão de dados para treino e teste
from sklearn.model_selection import train_test_split
#por padrão esta função separa 25% para teste e 75% para trienamento
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression

est = LinearRegression()

#treinamento e predição
est.fit(X_train, y_train)
y_pred = est.predict(X_test)
```

## Avaliando o Modelo

Como os targets são contínuos, uma alternativa é somar as diferenças entre o que foi estimado e o valor real, a fim de quantificar o erro da predição. Entretanto, podem aparecer valores positivos e negativas e ao somarmos não teríamos o erro não seria devidamente mensurado. Assim, ao invés de somar as diferenças, somamos o quadrado das diferenças e depois fazamos uma média. Essa técnica é chamada de Mean Squared Error (MSE).
<br>
Uma prática comum é calcular raiz quadrado do MSE, isto é Root Mean Squared Erro (RMSE), pois assim temos uma noção unitária do erro da função.

```python
import numpy as np
from sklearn.metrics import mean_squared_error

print('MSE: ', mean_squared_error(y_test, y_pred))
print('RMSE: ', np.sqrt(mean_squared_error(y_test, y_pred)))
```

## Interpretando os coeficientes do modelo

```python
print(est.intercept_)
print(est.coef_)
```

```python
#vamos deixar em pares os nomes dos atributos e os coeficientes obtidos pelo modelo
list(zip([['TV', 'Radio', 'Newspaper'], est.coef_]))
```

**O modelo obtido foi:**

$ y = 2.876966 + 0.04656457 \times TV + 0.17915812 \times Radio + 0.00345046 \times Newspaper $

How do we interpret the **TV coefficient** (0.0466)?
* For a given amount of Radio and Newspaper ad spending, a "unit" increase in TV ad spending is associated with a 0.0466 "unit" increase in Sales. * Or more clearly: For a given amount of Radio and Newspaper ad spending, an additional 1,000 spent on TV ads is associated with an increase in sales of 46.6 items.

Important notes:

* This is a statement of association, not causation.
* If an increase in TV ad spending was associated with a decrease in sales, $\beta_1$ would be negative.
